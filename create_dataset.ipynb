```python
def process_prompts(df, model, tokenizer, device):
    # Create a copy of the DataFrame to store results
    results_df = df.copy()
    
    # Initialize all responses as empty strings
    results_df['response'] = ''
    results_df['model'] = model.name_or_path
    
    # Process each row
    for idx in results_df.index:
        # Create message format for current prompt
        messages = [{"role": "user", "content": results_df.loc[idx, 'prompt']}]
        
        # Generate response
        try:
            input_text = tokenizer.apply_chat_template(messages, tokenize=False)
            inputs = tokenizer.encode(input_text, return_tensors="pt").to(device)
            outputs = model.generate(inputs, max_new_tokens=512, temperature=0.2, top_p=0.9, do_sample=True)
            response = tokenizer.decode(outputs[0])[len(input_text):].strip("<|im_start|>assistant").strip("<|im_end|>").strip("\n") # tolgo un po' di formattazione
            
            # Store the response in the current row
            results_df.loc[idx, 'response'] = response
            
            print(f"Processed row {idx}/{len(results_df.index)}. Answer: {response[:50]}")
            
        except Exception as e:
            print(f"Error processing row {idx}: {str(e)}")
            results_df.loc[idx, 'response'] = f"Error: {str(e)}"
    
    return results_df

# Usage:
results_df = process_prompts(df.loc[df.global_index < 139], model, tokenizer, device)
```

```python
import json

def df_to_json(df, output_file):
    # Convert DataFrame to list of dictionaries
    results = df.to_dict('records')
    
    # Convert numeric types to int
    for result in results:
        result['global_index'] = int(result['global_index'])
        result['index'] = int(result['index'])
        result['category_id'] = int(result['category_id'])
        
    
    # Save to JSON file
    with open(output_file, 'w') as f:
        json.dump(results, f, indent=4)
    
    return results

output_file = 'evaluation_samu.json'
results = df_to_json(results_df, output_file)
```
